```python
data = pd.read_csv("Clean_data.csv")
```

```python
data
```

```python
dummies=pd.get_dummies(data.location)
dummies.head(10)
```

```python
df=pd.concat([data,dummies.drop('other',axis='columns')],axis='columns')
df.head()
```

```python
df1 = df.drop('location',axis='columns')
df1.head(2)
```

```python
df1.shape
```

```python
X = df1.drop('price',axis='columns')
X.head()
```

```python
y=df1.price
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=36)
```

```python
ridge = Ridge()
ridge.fit(X_train, y_train)
rid_pred = ridge.predict(X_test)
rid_mse = mean_squared_error(y_test, rid_pred)
rid_r2 = r2_score(y_test, rid_pred)
print(rid_mse)
print(rid_r2)
```

```python
def predict_price(location,sqft,bath,BHK):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = BHK
    if loc_index >= 0:
        x[loc_index] = 1

    return ridge.predict([x])[0]
```
